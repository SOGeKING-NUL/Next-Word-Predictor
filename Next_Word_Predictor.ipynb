```python
import pandas as pd
import numpy as np
```

```python
snippet= pd.read_csv('/content/my_sentences.txt', header=None, delimiter= '/n')
```

```python
snippet
```

```python
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
texts = snippet[0].tolist()  # Convert the first column to a list
texts
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts) # Fit the tokenizer on the texts (this builds the word index)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
#this piece of code takes a sentence from texts, and then makes a sequence of it using the tokenized word index and then adds a seqentially increasing list of the words in the sentence into a list
input_sequences=[]

for sentence in texts:
    tokenized_sentence= tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)): #starts from 1 as the first word in the sentence has to be there from the start only
      input_sequences.append(tokenized_sentence[:i+1])




```

```python
input_sequences
```

```python
#since the length of all the sentences is not the same, padding is is needed

#finding the lenght of the longest sentence

max_sentence_len= max(len(x) for x in (input_sequences))


max(len(x) for x in (input_sequences))
```

```python
#padding

from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequence= pad_sequences(input_sequences, maxlen=max_sentence_len, padding='pre')
```

```python
padded_input_sequence[346]
```

```python
snippet= pd.DataFrame(padded_input_sequence)
```

```python
snippet
```

```python
# now we need to train the mode to predict the last word of the sequence

X=snippet.iloc[:, :-1]
y= snippet.iloc[:, -1]
```

```python
print(X.shape)
print(y.shape)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=2031) #classes is 2030+ 1 since the word_index is 2030 but the to_categorical starts from 0 and will thus miss the last word, we need to +1 to accomodate the last word in the word_index
```

```python
print(y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model= Sequential()
model.add(Embedding(input_dim=2031, output_dim=1000, input_length=22))# 2031 words, 1000 nodes layer to reduce to, 22 is the len of the longest sentence
model.add(LSTM(300,return_sequences=True))
model.add(LSTM(300))
model.add(Dense(2031, activation='softmax')) # since we have 2031 words
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=10)
```

```python
import time
text = "I appreciate"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
